## Step 2: Initialise the client

Once the library is installed, a Client object must be created. It is responsible for sending sending requests to the server and processing responses in the local environment. It enables a seamless interaction with the server. 

To create the client, one needs to give it a few parameters:
- a url: the root application endpoint to the remote secure server.
- user_name: her name as registered in the database (Jack)
- dataset_name: the name of the dataset that we want to query (TITANIC)

```python
API_URL = "http://lomas_server"
USER_NAME = "Alice"
DATASET_NAME = "PENGUIN"
client = Client(url=API_URL, user_name = USER_NAME, dataset_name = DATASET_NAME)
```

### Retrieve access token at http://lomas_server/token to access /get_dataset_metadata endpoint

```python
import requests
import json

def get_access_token(api_url: str, access_payload: dict):
    """Get JWT token.

    Gets a JWT token after user has been authenticated.
    :param api_url: URL of the API server.
    :param access_payload: Dictionary with keys 'username' and 'password' and
                            fitting values.
    :return: JWT Token
    """
    header = {
        "Accept-Encoding": "gzip, deflate, br",
        "Content-Type": "application/x-www-form-urlencoded;charset=UTF-8",
        "Accept": "*/*",
    }
    response = requests.post(api_url + "/token", headers=header, data=access_payload)
    return json.loads(response.content.decode("ascii"))["access_token"]
```

```python
access_payload = {
    "username": "alice",
    "password": "secret",
}
token = get_access_token(API_URL, access_payload)
token
```

### Retrieve resource after authorization has been granted

```python
titanic_metadata = client.get_dataset_metadata(token=token)
titanic_metadata
```
